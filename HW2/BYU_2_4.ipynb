```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w0 = 10
gamma = 1
w_p = 5 # Plasma frequency, assumed

w = np.linspace(0.01, 20 * w_p, 500) # Start from a small non-zero value

# Lorentz model for complex relative permittivity
epsilon_r = 1 + (w_p**2) / (w0**2 - w**2 - 1j * gamma * w)

epsilon_r_real = np.real(epsilon_r)
epsilon_r_imag = np.imag(epsilon_r)

# Calculate n and kappa
n_squared = 0.5 * (epsilon_r_real + np.sqrt(epsilon_r_real**2 + epsilon_r_imag**2))
kappa_squared = 0.5 * (-epsilon_r_real + np.sqrt(epsilon_r_real**2 + epsilon_r_imag**2))

n = np.sqrt(n_squared)
kappa = np.sqrt(kappa_squared)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(w / w_p, n, label='n (refractive index)')
plt.plot(w / w_p, kappa, label='kappa (extinction coefficient)')
plt.xlabel(r'$\omega / \omega_p$')
plt.ylabel('n or $\kappa$')
plt.title('Refractive Index and Extinction Coefficient (Lorentz Model)')
plt.legend()
plt.grid(True)
plt.xlim(0, 20)
plt.ylim(bottom=0)
plt.show()
```

```python
w01 = 10
gamma1 = 1
f1 = 0.5
w02 = 15
gamma2 = 1
f2 = 0.25
w03 = 25
gamma3 = 3
f3 = 0.25
```

```python
w = np.linspace(0.01, 30 * w_p, 500) # Frequency range from 0 to 30 * omega_p

# Lorentz model for complex relative permittivity with three resonances
epsilon_r1 = (f1 * w_p**2) / (w01**2 - w**2 - 1j * gamma1 * w)
epsilon_r2 = (f2 * w_p**2) / (w02**2 - w**2 - 1j * gamma2 * w)
epsilon_r3 = (f3 * w_p**2) / (w03**2 - w**2 - 1j * gamma3 * w)

epsilon_r_total = 1 + epsilon_r1 + epsilon_r2 + epsilon_r3

epsilon_r_real = np.real(epsilon_r_total)
epsilon_r_imag = np.imag(epsilon_r_total)

# Calculate n and kappa
n_squared = 0.5 * (epsilon_r_real + np.sqrt(epsilon_r_real**2 + epsilon_r_imag**2))
kappa_squared = 0.5 * (-epsilon_r_real + np.sqrt(epsilon_r_real**2 + epsilon_r_imag**2))

n = np.sqrt(n_squared)
kappa = np.sqrt(kappa_squared)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(w / w_p, n, label='n (refractive index)')
plt.plot(w / w_p, kappa, label='kappa (extinction coefficient)')
plt.xlabel(r'$\omega / \omega_p$')
plt.ylabel('n or $\kappa$')
plt.title('Refractive Index and Extinction Coefficient (Three Resonance Lorentz Model)')
plt.legend()
plt.grid(True)
plt.xlim(0, 30)
plt.ylim(bottom=0)
plt.show()
```
