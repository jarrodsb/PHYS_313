```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fused Silica
fsB1 = 6.96166300e-01
fsB2 = 4.07942600e-01
fsB3 = 8.97479400e-01
fsC1 = 4.67914826e-03
fsC2 = 1.35120631e-02
fsC3 = 9.79340023e+01

# Schott BK7
sbk7B1 = 1.03961212e+00
sbk7B2 = 2.31792344e-01
sbk7B3 = 1.01046945e+00
sbk7C1 = 6.00069867e-03
sbk7C2 = 2.00179144e-02
sbk7C3 = 1.03560653e+02

# Sapphire (Ordinary)
sphB1 = 1.43134930e+00
sphB2 = 6.50547130e-01
sphB3 = 5.34140210e+00
sphC1 = 5.27992610e-03
sphC2 = 1.42382647e-02
sphC3 = 3.25017834e+02
```

```python
def Sellmeier(wavelength, B1, B2, B3, C1, C2, C3):
  # Assuming wavelength is in micrometers
  wavelength_squared = wavelength**2
  n_squared_minus_1 = (B1 * wavelength_squared) / (wavelength_squared - C1) + \
   (B2 * wavelength_squared) / (wavelength_squared - C2) + \
    (B3 * wavelength_squared) / (wavelength_squared - C3)
  n_squared = 1 + n_squared_minus_1
  return np.sqrt(n_squared)
```

```python
x = np.linspace(0.4, 0.7, 300) # Visible spectrum ranging from 0.4 (blue) to 0.7 (red) microns
```

```python
# Plotting
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True, figsize=(10, 12))

# Plot Fused Silica
ax1.plot(x, Sellmeier(x,fsB1,fsB2,fsB3,fsC1,fsC2,fsC3), label='Fused Silica', color='blue')
ax1.set_ylabel('n (Fused Silica)')
ax1.set_title('Sellmeier Dispersion Equations')
ax1.legend()
ax1.grid(True)

# Plot Schott BK7
ax2.plot(x, Sellmeier(x,sbk7B1,sbk7B2,sbk7B3,sbk7C1,sbk7C2,sbk7C3), label='Schott BK7', color='green')
ax2.set_ylabel('n (Schott BK7)')
ax2.legend()
ax2.grid(True)

# Plot Sapphire
ax3.plot(x, Sellmeier(x,sphB1,sphB2,sphB3,sphC1,sphC2,sphC3), label=r'Sapphire $\eta_o$', color='red')
ax3.set_ylabel(r'n (Sapphire $\eta_o$)')
ax3.set_xlabel('Wavelength (microns)')
ax3.legend()
ax3.grid(True)

plt.tight_layout()
plt.show()
```
