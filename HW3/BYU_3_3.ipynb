```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta = np.linspace(0,90,90)

ni = 1

nt = 2.42
```

```python
# Snell's Law
def snell(ni, nt, theta_i):
  return np.arcsin(ni*np.sin(theta_i)/nt )

# Fresnel Coefficients
def rs(ni, nt, theta_i):
  theta_t = snell(ni, nt, theta_i)
  return (ni*np.cos(theta_i) - nt*np.cos(theta_t))/(ni*np.cos(theta_i) + nt*np.cos(theta_t))

def rp(ni, nt, theta_i):
  theta_t = snell(ni, nt, theta_i)
  return (ni*np.cos(theta_t) - nt*np.cos(theta_i))/(ni*np.cos(theta_t) + nt*np.cos(theta_i))

def ts(ni, nt, theta_i):
  theta_t = snell(ni, nt, theta_i)
  return (2*ni*np.cos(theta_i))/(ni*np.cos(theta_i) + nt*np.cos(theta_t))

def tp(ni, nt, theta_i):
  theta_t = snell(ni, nt, theta_i)
  return (2*ni*np.cos(theta_i))/(ni*np.cos(theta_t) + nt*np.cos(theta_i))
```

```python
# Plotting
plt.figure(figsize=(10,10))
plt.plot(theta, rs(ni, nt, np.radians(theta)), label=r'$r_s$', color='red', linestyle='--')
plt.plot(theta, ts(ni, nt, np.radians(theta)), label=r'$t_s$', color='blue', linestyle='--')
plt.plot(theta, rp(ni, nt, np.radians(theta)), label=r'$r_p$', color='red', linestyle='-')
plt.plot(theta, tp(ni, nt, np.radians(theta)), label=r'$t_p$', color='blue', linestyle='-')

# Calculate Brewster angle
theta_b = np.degrees(np.arctan(nt / ni))
plt.axvline(x=theta_b, color='black', linestyle='--', label='Brewster Angle')

plt.xlabel(r'$\theta_i$ (degrees)')
plt.ylabel('Coefficient')
plt.title(r'Fresnel Coefficients: $n_i=1$ (Air), $n_t=2.42$ (Diamond)')
plt.legend()
plt.grid(True)
plt.ylim(-1, 1.5)
plt.show()
```
